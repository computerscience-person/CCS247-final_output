## With numerical data processing

### Preprocessing

```python
from sklearn.preprocessing import OrdinalEncoder, StandardScaler, Normalizer
from sklearn.compose import ColumnTransformer
```

```python
dataset_churn.describe()
```

```python
preprocessor = ColumnTransformer(
    [
        ("CreditScoreScaled", StandardScaler(), ["CreditScore"]),
        ("GeographyEncoded", OrdinalEncoder(), ["Geography"]),
        ("GenderEncoded", OrdinalEncoder(), ["Gender"]),
        ("EstimatedSalaryScaled", StandardScaler(), ["EstimatedSalary"]),
        ("BalanceScaled", StandardScaler(), ["Balance"]),
    ],
    remainder="passthrough"
)
preprocessor.set_output(transform="polars")
```

```python
dataset_churn_processed = preprocessor.fit_transform(dataset_churn.drop(["#", "CustomerId", "Surname"]).collect())
```

```python
dataset_churn_processed.head(15)
```

```python
dataset_churn_processed.corr().hvplot.heatmap(colorbar=True, height=750, width=750, symmetric=True, xaxis=None, yaxis=None)
```

```python
dataset_churn_processed.select([
    "CreditScoreScaled__CreditScore",
    "EstimatedSalaryScaled__EstimatedSalary",
    "BalanceScaled__Balance"
    ]).hvplot.kde()
```

The CreditScore looks to have a normal distribution; meanwhile, EstimatedSalary looks to have a uniform distribution, and Balance has a bimodal distribution. 

```python
dataset_churn_processed.describe()
```

Looking at the correlation heatmaps, customer metrics don't seem to correlate much with customer churn.
