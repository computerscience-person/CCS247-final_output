# Importing some stuff

```python
import polars as pl
import hvplot.polars
```

# Inital dataset querys

```python
dataset_churn = pl.scan_csv('Churn_Modelling.csv')
dataset_churn.head().collect()
```

```python
dataset_churn.select(pl.all().n_unique()).collect()
```

```python
dataset_churn.describe()
```

```python
dataset_churn_numerical = dataset_churn.drop(pl.selectors.string())
dataset_churn_numerical.describe()
```

```python
dataset_churn_numerical.collect().corr()
```

```python
dataset_churn_numerical.collect().corr().hvplot.heatmap(colorbar=True, height=750, width=750, symmetric=True, xaxis=None, yaxis=None)
```

According to the correlation heatmap, the (numerical) columns of the dataset does not show much correlation with each other.
